# O(1) 时间插入、删除和获取随机元素 - 允许重复

设计一个支持在平均 时间复杂度 O(1) 下， 执行以下操作的数据结构。

注意: 允许出现重复元素。

insert(val)：向集合中插入元素 val。
remove(val)：当 val 存在时，从集合中移除一个 val。
getRandom：从现有集合中随机获取一个元素。每个元素被返回的概率应该与其在集合中的数量呈线性相关。
示例:

```python
// 初始化一个空的集合。
RandomizedCollection collection = new RandomizedCollection();

// 向集合中插入 1 。返回 true 表示集合不包含 1 。
collection.insert(1);

// 向集合中插入另一个 1 。返回 false 表示集合包含 1 。集合现在包含 [1,1] 。
collection.insert(1);

// 向集合中插入 2 ，返回 true 。集合现在包含 [1,1,2] 。
collection.insert(2);

// getRandom 应当有 2/3 的概率返回 1 ，1/3 的概率返回 2 。
collection.getRandom();

// 从集合中删除 1 ，返回 true 。集合现在包含 [1,2] 。
collection.remove(1);

// getRandom 应有相同概率返回 1 和 2 。
collection.getRandom();
```

````python
from collections import defaultdict
import random

class RandomizedCollection:

    def __init__(self):
        """
        Initialize your data structure here.
        """
        self.numPosIdx = defaultdict(list)
        self.nums = []
        

    def insert(self, val):
        """
        Inserts a value to the collection. Returns true if the collection did not already contain the specified element.
        :type val: int
        :rtype: bool
        """
        exists = self.numPosIdx[val]
        self.nums.append((val, len(exists)))
        exists.append(len(self.nums) - 1)
        if (len(exists) == 1):
            return True
        return False
        
        

    def remove(self, val):
        """
        Removes a value from the collection. Returns true if the collection contained the specified element.
        :type val: int
        :rtype: bool
        """
        exists = self.numPosIdx[val]
        
        if len(exists) == 0:
            return False
        
        lastPosOfNum = exists.pop()
        lastInArr = self.nums.pop()
        
        if (lastPosOfNum == len(self.nums)):
            return True
        
        self.nums[lastPosOfNum] = lastInArr
        self.numPosIdx[lastInArr[0]][lastInArr[1]] = lastPosOfNum
        return True
        

    def getRandom(self):
        """
        Get a random element from the collection.
        :rtype: int
        """
        return self.nums[int(random.random() * len(self.nums))][0]
        


# Your RandomizedCollection object will be instantiated and called as such:
# obj = RandomizedCollection()
# param_1 = obj.insert(val)
# param_2 = obj.remove(val)
# param_3 = obj.getRandom()
````

````python
def doSequence(sequence):
    rc = RandomizedCollection()
    result = []
    for s in sequence:
        if s[0] == 'i':
            result.append(rc.insert(s[1]))
        elif s[0] == 'r':
            result.append(rc.remove(s[1]))
    return result
            
            
````

````python
def test():
    r = RandomizedCollection()
    randomResults = []
    for i in range(10):
        r.insert(0)
        
    for i in range(590):
        r.insert(1)
        
    for i in range(550):
        r.insert(2)
        
    for i in range(200):
        removeResult = r.remove(2)
        assert removeResult == True
        
    for i in range(100):
        r.insert(1)
    
    for i in range(100):
        r.insert(2)
        
    for i in range(150):
        r.remove(2)
        
    removeResult = r.remove(100)
    assert removeResult == False
    
    for i in range(1000000):
        randomResults.append(r.getRandom())
        
    assert 9000 <= len(list(filter(lambda n: n == 0, randomResults))) <= 11000
    assert 680000 <= len(list(filter(lambda n: n == 1, randomResults))) <= 700000
    assert 290000 <= len(list(filter(lambda n: n == 2, randomResults))) <= 310000
    
    assert str(doSequence([('i', 1), ('i', 1), ('i', 2), ('i', 2), ('i', 2), ('r', 1), ('r', 1), ('r', 2), ('i', 1), ('r', 2)])) == str([True, False, True, False, False, True, True, True, True, True])
    
    print("all passed")
````

````python
test()
````
